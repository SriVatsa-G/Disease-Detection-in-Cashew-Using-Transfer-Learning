```python
from tensorflow.keras.models import load_model
import numpy as np
from tensorflow.keras.preprocessing import image
import os

# Load the saved model
model = load_model("/mnt/c/SV-Desktop/FALL SEMESTER 2024-25/Dissertation 1/Implementation/Models/Raw/EfficientNet/best_model_B4.keras")
```

```python
# Ensure that class labels match the ones used during training
class_labels = {'anthracnose': 0, 'healthy': 1, 'leaf miner': 2, 'red rust': 3}
class_indices = {v: k for k, v in class_labels.items()}  # Reverse the dictionary to map indices back to class names

# Function to preprocess a new image (consistent with training preprocessing)
def preprocess_image(img_path, target_size=(224, 224)):
    img = image.load_img(img_path, target_size=target_size)  # Load image with the target size
    img_array = image.img_to_array(img)  # Convert image to a numpy array
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    img_array = img_array / 255.0  # Normalize the image to the range [0, 1] (same as during training)
    return img_array

# Function to load model and predict the class
def predict_class(img_path):
    # Preprocess the image
    preprocessed_img = preprocess_image(img_path)
    
    # Predict using the loaded model
    predictions = model.predict(preprocessed_img)
    
    # Get the index of the predicted class
    predicted_class_index = np.argmax(predictions, axis=1)[0]
    
    # Get the class label from the predicted index
    predicted_class_label = class_indices[predicted_class_index]
    
    # Print the predicted class
    print(f"Predicted class: {predicted_class_label} (Index: {predicted_class_index})")
    return predicted_class_label
```

```python
# Path to the test image
img_path = '/mnt/c/SV-Desktop/FALL SEMESTER 2024-25/Dissertation 1/Implementation/testing/leaf miner53_.jpg'

# Predict and print the class of the test image
predicted_class = predict_class(img_path)
```
